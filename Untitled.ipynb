```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from fealpy.mesh import TriangleMesh
```

```python
node = np.array([
        (0, 0),
        (1, 0),
        (1, 1),
        (0, 1)
    ], dtype=np.float)
cell = np.array([
        (1, 2, 0),
        (3, 0, 2)
    ], dtype=np.int)
```

```python
mesh = TriangleMesh(node, cell)
mesh.uniform_bisect(4)
```

```python
fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes)
```
