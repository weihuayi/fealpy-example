```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from fealpy.mesh import TriangleMesh
```

```python
node = np.array([
        (0, 0),
        (1, 0),
        (1, 1),
        (0, 1)
    ], dtype=np.float)
cell = np.array([
        (1, 2, 0),
        (3, 0, 2)
    ], dtype=np.int)
```

```python
mesh = TriangleMesh(node, cell)
mesh.uniform_refine(4)
```

```python
fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes)
```

```python
from fealpy.pde.poisson_1d import CosData
```

```python
pde = CosData()
mesh = pde.init_mesh(n=4)
node = mesh.entity('node')
cell = mesh.entity('cell')
print(type(mesh))
print(node)
print(cell)
fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes, markersize=20)
mesh.find_node(axes, markersize=20, showindex=True, fontsize=12)
```

```python
uI = pde.solution(node)
print(uI)
```

```python
from fealpy.mesh import StructureIntervalMesh

smesh = StructureIntervalMesh([0, 1], 0.2)
print(type(smesh))
fig = plt.figure()
axes = fig.gca()
smesh.add_plot(axes, markersize=10)
smesh.find_node(axes, markersize=10)
```

```python
node = smesh.entity('node')
uI = pde.solution(node)
plt.plot(node, uI)
```

```python
from scipy.sparse.linalg import spsolve
from scipy.sparse import csr_matrix

def u(x):
    return np.sin(np.pi*x)

def f(x):
    return np.pi**2*np.sin(np.pi*x)

smesh = StructureIntervalMesh([0, 1], 0.025)
NN = smesh.number_of_nodes()

A = smesh.laplace_operator().tocsr()

node = smesh.entity('node')
b = f(node)

uh = np.zeros(NN, dtype=np.float)
uh[1:-1] = spsolve(-A[1:-1, :][:, 1:-1], b[1:-1])

error = np.max(np.abs(uh - u(node)))
print(error)
```
